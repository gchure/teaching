In this tutorial, we will investigate the dynamics of mRNA expression from an unregulated (constitutive) promoter.

## The dynamics of constitutive expression

Let's imagine that we have a collection of bacterial cells. In their genomes, there is a gene which is constitutively expressed meaning that there is no other regulation of expression other than competing with the rest of the genome for RNA polymerase molecules. What is the mean number of mRNA molecules per cell as a function of time? To answer this question, we have to have some knowledge of rates. The most obvious value we need to figure out is how rapidly are mRNA transcripts made from our promoter (the production rate). Are mRNAs molecules stable over time, or do they degrade? If they degrade, what is the rate of this degradation? As is usually true, we can draw a cartoon of our gene of interest and and explicitly state what our parameters are. 

![](../figures/constitutive_expression.png)

In this cartoon, we've denoted the production rate as $r$ and the degradation rate as $\gamma$. It is important to note here that the production rate is dependent only on the availability of RNA polymerase molecules to perform transcription. The degradation rate is also constant for a single mRNA, however the number of molecules degraded per unit time is dependent on how many mRNAs are present in the cell.

The change in mRNA expression with respect to time is given by

$$
\frac{dm}{dt} = r - \gamma m,
$$

where $r$ is the production rate of mRNA, $\gamma$ is the degradation rate of mRNA, and $m$ is the number of mRNA molecules. This means that the production of mRNA is the same regardless of the number of mRNA molecules that are currently in the cell. The degradation component, however, **is** dependent on the concentration of mRNA. Before we begin our integration, let's examine how each of these components contribute to expression independently. We'll first define some model parameters and then plot their values over a range of mRNA molecules.

```python
# Assign values to our model parameters.
r = 1  # mRNA production rate in 1/min.
gamma = 1/3  # mRNA decay rate in 1/min.
num_mRNA = np.arange(0, 10, 1)  # Set up a vector of mRNAs from 0 to 10

# Determine the contribution of each component.
prod = np.ones_like(num_mRNA) * r  # Vector of production component.
deg = gamma * num_mRNA  # Vector of degradation component.

# Plot the two together.
plt.plot(num_mRNA, prod, '-', label='production')
plt.plot(num_mRNA, deg, '-', label='degradation')

# Add a legend and axis labels.
plt.legend(loc='upper left')
plt.xlabel('number of mRNA')
plt.ylabel(r'contribution to $dm/dt$')
```

What happens at the intersection? At this point, the production component is equal to the degradation component, meaning this is the **equilibrium steady state** of the system. How long does it take to reach equilibrium? To find out, let's try out some numerical integration. We can solve for the steady state of the system by setting our differential equation to zero,

$$
\frac{dm}{dt} = r - \gamma m = 0.
$$

Solving for $m$ yields

$$
m = \frac{r}{\gamma}.
$$

Plugging in the values we defined above, we find that

$$
\frac{r}{\gamma} = \frac{1}{\frac{1}{3}} = 3,
$$

which is precisely the point of intersection in our plot above! How long does it take to reach equilibrium? To find out, let's perform some numerical integration.


## Euler integration

We can integrate this differntial equation analytically, but numerical integration is a very useful skill to have in your toolbox. In this tutorial, we will use [Euler integration](https://en.wikipedia.org/wiki/Euler_method). In short, our algorithm will be as follows:

1. Compute the change in number of mRNA molecules at a give time step $\Delta t$.
2. Add this change in mRNA copy number to the number of molecules computed in the previous time step. 
3. Take a step forward in time and repeat. 

This same procedure will be repeated over a large number of time steps. To be more explicit, at time step $i$, we will evaluate 

$$
m_{t=i} = m_{t= i - 1} + r\Delta t - \gamma m_{t= i - 1}\Delta t.
$$

where $\Delta t$ is the time difference between $t=i$ and $t = i - 1$.

We have a few more parameters we have to define for our integration. Since we are integrating over time, we have to specify the timescale over which we will integrate. We also have to define how many points over this time scale we would like to integrate. Finally, we have to define our initial conditions. We'll first assume we are a cell with 0 copies of the mRNA of interest.   


```python
# Define the parameters of our integration
time = 20  # in units of minutes.
dt = 0.1  # time step in units of min
num_steps = int(time / dt)  # the number of time steps to integrate.
time_vec = np.linspace(0, time, num_steps)  # Define the time vector.
init_condition = 0  # number of mRNA molecules at time 0.
```

Note that in the above code cell, we converted the `num_steps` variable to a float. we did this because we want to take a finite number of steps and Python 3.5 division evaluation returns a float. With these values defined, we can now perform the integration. We will first make an empty vector where we will store the number of mRNA molecules at each point in time and then compute the value of our differential equation at each time step.  

```python
# Set up a storage vector for our mRNA and set the initial condition.
m_t = np.empty_like(time_vec)
m_t[0] = init_condition

# Loop through each time point and evaluate the differential equation.
for i in range(1, num_steps):
    m_t[i] = m_t[i-1] + r*dt - gamma * m_t[i-1]*dt 
```

Now, let's plot the trajectory. We'll also plot the steady state value of $r/y = 3$ to verify our integration. 

```python
# Plot the result of our integration.
plt.plot(time_vec, m_t, '-', label='integration')

# Plot the steady state as a horizontal line.
plt.hlines(r/gamma, xmin=0, xmax=time, label='steady state')

# Extend the margins and add axis labels.
plt.ylim([0, 5])
plt.xlabel('time (min)')
plt.ylabel(r'$m(t)$')
plt.legend()
```

We can see that after about 15-ish minutes, the system has reached the equilibrium steady state. What happens when we change our initial condition? Let's consider the case where there are 10 mRNAs to begin with in the cell. 

```python
# Set up a storage vector for our mRNA and set the new initial condition.
m_t2 = np.empty_like(time_vec)
m_t2[0] = 10  # Our new initial condition

# Loop through each time point and evaluate the differential equation.
for i in range(1, len(time_vec)):
    m_t2[i] = m_t2[i-1] + r*dt - gamma * m_t2[i-1]*dt 
    
# Now we'll plot the two starting conditions and the steady state on the same axes.
plt.plot(time_vec, m_t, '-', label=r'$m_0 = 0$')
plt.plot(time_vec, m_t2, '-', label=r'$m_0 = 10$')
plt.hlines(r/gamma, xmin=0, xmax=time, label='steady state')

# Change the axes limits and add labels.
plt.ylim([0, 12])
plt.xlabel('time (min)')
plt.ylabel(r'$m(t)$')
plt.legend()
```

As expected, the number of mRNAs *decreased* with time because we were above the equilibrium value.
